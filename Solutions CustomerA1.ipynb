```python
import pandas as pd 
import numpy as np 
```

```python
file1 = pd.read_csv("csv_files/file1.csv")
file2 = pd.read_csv("csv_files/file2.csv")
file3 = pd.read_csv("csv_files/file3.csv")

df = pd.concat([file1, file2, file3], axis = 0)

display(df.shape)
```

```python
display(df.columns)
```

```python
cols = []
for i in range(len(df.columns)):
    cols.append(df.columns[i].lower())
df.columns = cols 
display(cols)
```

```python
display(df.head())
```

```python
df = df.drop(['education', 'number of open complaints'], axis = 1)
df.head()
```

```python
#Correct the values in the column customer lifetime value. 
#They are given as a percent, so multiply them by 100 and change dtype to numerical type.
```

```python
df.dtypes
```

```python
df['customer lifetime value'] = pd.to_numeric(df['customer lifetime value'], errors = 'coerce')
```

```python
df['customer lifetime value'].head()
```

```python
display(df.head())
```

```python
#Check for duplicate rows in the data and remove if any.
```

```python
df = df.drop_duplicates().reset_index(drop = True)

display(df.head())
```

```python
#Filter out the data for customers who have an income of 0 or less
```

```python
df = df[df['income'] > 0].reset_index(drop = True)

df.head()
```
